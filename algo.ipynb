```
def write_to_csv(y):
  df = pd.DataFrame(columns=['id','label'])
  for id,x in enumerate(y):
    df = df.append({'id':id+30000,'label':int(x)},ignore_index=True)  
  df.to_csv("result.csv",index=False)
```
